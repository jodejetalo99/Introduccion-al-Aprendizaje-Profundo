```python
class UTKFace:

    def __init__(self, root, subset, split=0.80, seed=0, transform=None):
        # guardamos atributos
        self.root = join(root, 'UTKFace')
        # proporcion de los datos de entrenamiento
        self.split = split
        # subconjunto de datos
        self.subset = subset
        # si aplicamos transformacion
        self.transform = transform
        
        imagenes = os.listdir(self.root)
        # fijamos el barajeo para reproducibilidad
        random.Random(seed).shuffle(imagenes) 
        # obtenemos la edad de cada participante
        edades = []
        for im in imagenes:
          edades.append(int(im.split("_")[0]))

        # hacemos la particion de datos en entrenamiento y prueba
        tamanio_trn = int(len(imagenes)*self.split) 
        if self.subset == "trn":
            self.imagenes = imagenes[:tamanio_trn]
            self.edades = edades[:tamanio_trn]
        elif self.subset == "tst":
            self.imagenes = imagenes[tamanio_trn:]
            self.edades = edades[tamanio_trn:]
        # subconjuntos solamente de entrenamiento y prueba
        else:
          ValueError(f'invalid value for subset={subset}')


    def __getitem__(self, i):
        # cargamos la imagen
        imagen = self.imagenes[i]
        img = Image.open(join(self.root,imagen))
        #print(self.root,imagen)
        # aplicamos transformaciÃ³n
        if self.transform is not None:
            img = self.transform(img)
        edad = self.edades[i]
 
        return {'x': img, 'edad': edad}

    def __len__(self):
        # regresamos numeros de ejemplos
        return len(self.imagenes)
```

### 3.1 Carga de pesos

```python
# instancia de modelo y carga de pesos
model = tvm.shufflenet_v2_x0_5(pretrained=True)
model
```

Podemos observar que el nÃºmero de caracterÃ­sticas en la Ãºltima capa es: 1024. AdemÃ¡s, necesitamos que sea un regresor, no un clasificador, por ello, hacemos las respectivas modificaciones.



```python
class Modelo(nn.Module):
    
    def __init__(self, pretrained):
        super(Modelo, self).__init__()
        
        # cargamos el modelo
        self.cnn = tvm.shufflenet_v2_x0_5(pretrained=pretrained)
        # remplazamos la FC para dejar intactas 
        # las caracterÃ­sticas (diablito)
        self.cnn.fc = nn.Identity()

        in_features = 1024
        # armamos la capa de regresiÃ³n
        self.reg = nn.Sequential(
            nn.BatchNorm1d(in_features),
            nn.Linear(in_features, 1)
        )
    
    def forward(self, x):
        # [N, 3, 200, 200] => [N, 1024]
        x = self.cnn(x)
        # [N, 1024] => [N, 1]
        x = self.reg(x)
        return x
    
model = Modelo(True)

# congelado de parÃ¡metros de la CNN
for param in model.cnn.parameters():
    param.requires_grad = False

model
```

### 4.1 Funciones de entrenamiento

```python
def train_epoch(dl, model, opt, device):
    """Entrena una Ã©poca"""
    # entrenamiento de una Ã©poca
    for x, y_true in trn_dl:
        x = batch['x'].to(device)
        y_true = batch['edad'].to(device)
        # computamos la prediccion
        y_pred = model(x)
        # computamos la pÃ©rdida
        loss = F.mse_loss(y_pred, y_true.view(-1,1).float())
        # vaciamos los gradientes
        opt.zero_grad()
        # retropropagamos
        loss.backward()
        # actualizamos parÃ¡metros
        opt.step()
```

```python
def eval_epoch(dl, model, device, num_batches=None):
    """Evalua una Ã©poca"""
    # evitamos que se registren las operaciones 
    # en la grÃ¡fica de cÃ³mputo
    with torch.no_grad():
        losses, R2 = [], []
        # validaciÃ³n de la Ã©poca con num_batches
        # si num_batches==None, se usan todos los lotes
        for batch in take(dl, num_batches):
            x = batch['x'].to(device)
            y_true = batch['edad'].to(device)
            # computamos la prediccion
            y_pred = model(x)
            
            # computamos la pÃ©rdida
            loss =F.mse_loss(y_pred, y_true.view(-1,1).float())
            # computamos la r^2
            r2 = r2_score(y_true.cpu(),y_pred.cpu())

            # guardamos histÃ³ricos
            losses.append(loss.item())
            R2.append(r2)

        loss = np.mean(losses)
        R2_F = np.mean(R2)
        
        return loss, R2_F
```

```python
def save_check_point(model, epoch, run_dir):
    """Guarda un punto de control."""
    torch.save(
        {
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
        }, 
        join(run_dir, 'weights.pth')
    )
```

```python
def train(model, trn_dl, tst_dl, epochs, trn_batches=None, tst_batches=None, ultima_capa=True):
    # optimizador
    opt = optim.Adam(model.parameters(), lr=1e-3)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    if ultima_capa:
        # modelo en modo de evaluaciÃ³n
        # para transferencia de conocimiento ademÃ¡s de congelar los pesos, 
        # es importante congelar las estadÃ­sticas Ïƒ y Î¼ 
        model.eval()

    # ciclo de entrenamiento
    best_r2 = 0
    losses, R2 = {"trn":[], "tst":[]}, {"trn":[], "tst":[]} 
    for epoch in trange(epochs):

        # entrenamos la Ã©poca
        train_epoch(trn_dl, model, opt, device)

        # evaluamos la Ã©poca en entrenamiento
        trn_loss, trn_r2 = eval_epoch(trn_dl, model, device,  trn_batches)
        losses["trn"].append(trn_loss)
        R2["trn"].append(trn_r2)
        # evaluamos la Ã©poca en validaciÃ³n
        tst_loss, tst_r2 = eval_epoch(tst_dl, model, device, tst_batches)
        losses["tst"].append(tst_loss)
        R2["tst"].append(tst_r2)

        # pseudo paro temprano: si hay mejora guardamos punto de control
        if tst_r2 > best_r2:
            best_r2 = tst_r2
            save_check_point(model, epoch, run_dir)
```
